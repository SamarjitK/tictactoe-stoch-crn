```python
def clock_plot(minutes, fig):
    fig.clear()
    m = len(minutes)
    w = 2 * np.pi / m
    centers = [w * i for i in range(m)]
    ax = fig.add_subplot(111, polar=True)
    ax.set_xticklabels([str(i) for i in range(m)])
    ax.set_xticks(centers)
    ax.set_ylim(1/2, sim.simulator.n)
    ax.set_rorigin(- sim.simulator.n / 2)
    ax.bar(centers, minutes, width = w, bottom = 0.5, color = '0.8', edgecolor = 'k')
    plt.show()
    
    # TODO: get this to work for logscale
```

```python
def get_one_field(df, field):
    return df.transpose().groupby(level = field).sum().transpose()

def plot(df):
    fig = log_and_linear_plots(df)
    show(fig)

def get_multiline_fig(df, title = '', y_scale = 'linear'):
    # data-frame df has index of parallel time, columns giving counts of various states
    numlines=len(df.columns)
    p = figure(x_axis_label = "Parallel Time", y_axis_label = "Count", sizing_mode='stretch_width',  height = 400,  y_axis_type = y_scale)
    for i, col in enumerate(df.columns):
        color = Category10[10][i % 10]
        l = p.line(df.index, df[col] + 10 ** -2, line_width = 2, color = color, muted_color = color, muted_alpha = 0.2, legend_label = str(col))
        l.muted = len(df.columns) > 6
    p.legend.location = "top_right"
    p.legend.click_policy = "mute"
    p.legend.orientation = 'horizontal'
    p.title.text = title
    return p

def log_and_linear_plots(df, method = get_multiline_fig, title = ''):
    tabs = [Panel(child = method(df, title, y_scale), title = y_scale) for y_scale in ['linear', 'log']]
    return Tabs(tabs = tabs)

def plot_fields(df):
    figs = []
    for field in df.columns.names:
        figs.append(log_and_linear_plots(get_one_field(df, field), title=f'field: {field}'))
    show(layout(figs, sizing_mode = 'stretch_width'))
```
